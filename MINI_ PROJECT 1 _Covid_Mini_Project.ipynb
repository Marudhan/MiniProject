***1.Top 10 states having highest total death in the year 2020***

***7.Print list of Top N states by the least number of death in a specific year. Also needed to be ranked.***

A.Top 10 least death states in 2020

```python
#7.Print list of Top N states by the least number of death in a specific year. Also needed to be ranked.
#A.Top 10 least death states in 2020
def get_year(row):
    return datetime.strptime(row['date'], '%Y-%m-%d').year

df3['year'] = df3.apply(get_year, axis=1)
#filter on 2020
df_2020 = df3.where(df3['year'] == 2020)
df_top10_2020 = df_2020.groupby(['state'])['deaths'].sum().to_frame('count').reset_index()
df_top10_2020 = df_top10_2020.sort_values(by=['count'], ascending=True)



def new_rank(row):
    return int( 56 - row['rank1'])

df_top10_2020['rank1'] = df_top10_2020['count'].rank(ascending = False )
df_top10_2020['rank'] = df_top10_2020.apply(new_rank, axis=1)
df_top10_2020_sel = df_top10_2020[['state', 'count', 'rank']]
df_top10_2020_sel.head(10)
```
